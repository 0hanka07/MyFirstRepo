```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

# randomly chosen dataset
X = np.array([[3.5, 0.5], [3.2, 1.5], [3.0, 1.28],[1.5, 1.0], [1.35, 1.18], [2.0, 0.9],[3.0, 3.4], [3.10, 3.25], [3.1, 3.4]])

# Perform hierarchical clustering
dendrogram = dendrogram(linkage(X, method='ward'))

model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
model.fit(X)
labels = model.labels_
print(labels)
```

```python
plt.scatter(X[labels==0, 0], X[labels==0, 1], s=50, marker='o', color='red')
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=50, marker='*', color='blue')
plt.scatter(X[labels==2, 0], X[labels==2, 1], s=50, marker='+', color='green')
plt.show()
```
