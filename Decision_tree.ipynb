```python
# prompt: replace Iris dataset with breast cancer dataset

from sklearn.tree import plot_tree
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

# Load the Breast Cancer dataset
breast_cancer = load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target

# Print the first few rows of the dataset
print(X[:5])

# Print the target names
print(breast_cancer.target_names)

# To show the Breast Cancer dataset as a pandas DataFrame:
df = pd.DataFrame(data=np.c_[breast_cancer['data'], breast_cancer['target']],
                  columns=breast_cancer['feature_names'] + ['target'])
df['diagnosis'] = pd.Categorical.from_codes(breast_cancer.target, breast_cancer.target_names)
df.head()

plt.figure()
clf = DecisionTreeClassifier().fit(X, y)
plot_tree(clf, filled=True)
plt.title("Decision tree trained on the breast cancer dataset")
plt.show()
```
